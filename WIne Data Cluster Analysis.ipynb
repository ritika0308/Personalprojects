```python
from sklearn.cluster import DBSCAN
from collections import Counter
```

```python
wine_dbscan = DBSCAN(eps=25, min_samples = 3).fit(wine_data.iloc[:,1:14])
#performing DBSCAN with min pts = 3 and epsilon = 25
```

```python
wine_data2= wine_data
wine_data2['DBSCAN Labels']=wine_dbscan.labels_ #adding the lables as a column in the dataframe
print(Counter(wine_dbscan.labels_)) #counting the number of values in each label
```

```python
wine_data2 #displaying the dataframe
```

```python
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
wine_data_normalized = normalize(wine_data) #normalizing the wine data before performing Principal Component Analysis
pca_winedata = PCA(3) #reducing the dimensionality to trivariate data
pca_winedata.fit(wine_data_normalized)
  
pca_winedf = pd.DataFrame(pca_winedata.transform(wine_data_normalized)) #adding the reduced data into a dataframe
  
print(pca_winedf.head())
```

```python
print("Total explained variance ratio of trivariate data is: ",pca_winedata.explained_variance_ratio_[0]+pca_winedata.explained_variance_ratio_[1]+pca_winedata.explained_variance_ratio_[2])
```

```python
#for generating an array of colors for the 7 clusters(-1,0,1,2,3,4,5) we have which will be used in the 3D Plot
from matplotlib import colors as mcolors
import math
clusters = 7
colors = list(zip(*sorted((
                    tuple(mcolors.rgb_to_hsv(
                          mcolors.to_rgba(color)[:3])), name)
                     for name, color in dict(
                            mcolors.BASE_COLORS, **mcolors.CSS4_COLORS
                                                      ).items())))[1]
skips = math.floor(len(colors[30 : -35])/clusters)
cluster_colors = colors[30 : -35 : skips]

cluster_colors
```

```python
#generates a 3D scatter plot where each data point has a color related to its corresponding cluster
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib as mpl

#mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure(figsize=(32,20))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(pca_winedf[0], pca_winedf[1], pca_winedf[2], 
           c = list(map(lambda label : cluster_colors[label],
                                            wine_dbscan.labels_)))

str_labels = list(map(lambda label:'% s' % label, wine_dbscan.labels_))
   
list(map(lambda data1, data2, data3, str_label:
        ax.text(data1, data2, data3, s = str_label, size = 10,
        zorder = 20, color = 'k'), pca_winedf[0], pca_winedf[1],
        pca_winedf[2], str_labels))

plt.show()
```
